```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("train.csv")
```

```python
data.head()
```

```python
hours = pd.to_datetime(data["pickup_datetime"]).dt.strftime("%H")
```

```python
plt.figure(figsize=(15, 3))
sns.histplot(hours.sort_values(), discrete=True)
plt.show()
```

```python
distribution = hours.value_counts()
# abc_scaled = (abc_vc - abc_vc.min()) / (abc_vc.max() - abc_vc.min())
distribution_scaled = (distribution / distribution.max()) ** 2
```
